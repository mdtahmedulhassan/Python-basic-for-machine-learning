### Python OOP — Classes, Inheritance & Encapsulation

Covers the core OOP concepts:
- defining classes and methods
- `__init__` constructor
- class vs instance variables
- `getattr`, `setattr`, `hasattr`, `delattr`
- inheritance (single, multilevel, multiple)
- `super()`
- encapsulation (private attributes with `__`)
- nested (inner) classes

#### Basic class with methods

```python
class Calculator:
    def add(self, a, b):
        return a + b

    def subtract(self, a, b):
        return a - b

c1 = Calculator()
print(c1.add(10, 15))
print(c1.subtract(10, 15))
```

Methods can also save values to `self` so other methods can reuse them:

```python
class Calculator:
    def add(self, a, b):
        self.num1 = a
        self.num2 = b
        return self.num1 + self.num2

    def subtract(self):
        return self.num1 - self.num2   # reuses values stored by add()

c1 = Calculator()
print(c1.add(10, 15))
print(c1.subtract())
```

One method calling another inside the same class using `self`:

```python
class Calculator:
    def multiply(self, a, b):
        return a * b

    def subtract(self, a, b):
        return a - b

    def smart_calc(self, a, b):
        # if a is even → multiply, else → subtract
        if a % 2 == 0:
            return self.multiply(a, b)
        else:
            return self.subtract(a, b)

c = Calculator()
print(c.smart_calc(11, 5))   # 11 is odd → subtract
print(c.smart_calc(10, 5))   # 10 is even → multiply
```

#### Class variables vs instance variables

```python
class Player:
    sport = "Cricket"   # class variable — shared by all instances

    def show(self):
        print(self.sport)

p1 = Player()
p1.show()
```

Class can also read global variables (but prefer `self.` for clarity):

```python
name = "Virat"
age = 35

class Player:
    def show(self):
        print(name, age)   # reads from global scope

p1 = Player()
p1.show()
```

#### `__init__` constructor
`__init__` runs automatically when you create an object. Use it to set instance data.

```python
class Player:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def get_name(self):
        return self.name

    def get_age(self):
        return self.age

p1 = Player("Virat", 35)
print(p1.get_name())
print(p1.get_age())
```

#### Built-in attribute functions
`getattr`, `setattr`, `delattr`, `hasattr`

```python
class Player:
    def setup(self, name, age):
        self.name = name
        self.age = age

p1 = Player()
p1.setup("Sangakkara", 46)

print(getattr(p1, "age"))        # get value of an attribute by name
setattr(p1, "name", "Lara")      # change attribute value
delattr(p1, "name")              # delete an attribute
print(hasattr(p1, "age"))        # check if attribute exists -> True
```

#### Inheritance
A child class gets all methods of the parent class automatically.

**Single inheritance:**

```python
class Car:
    def info(self, model, year):
        self.model = model
        self.year = year
        return "Car info:", self.model, self.year

class Engine(Car):              # inherits Car
    def engine_model(self, num):
        self.num = num
        return self.num

c1 = Engine()
print(c1.engine_model("e160"))   # own method
print(c1.info("Corolla", 1997))  # inherited from Car
```

**Multilevel inheritance** — chain of classes:

```python
class Car:
    def info(self, model, year):
        self.model = model
        self.year = year
        return "Car info:", self.model, self.year

class Engine(Car):
    def engine_model(self, num):
        self.num = num
        return self.num

class Chassis(Engine):          # inherits Engine, which inherits Car
    def chassis_num(self, val):
        self.val = val
        return self.val

c1 = Chassis()
print(c1.chassis_num("A1235"))   # own method
print(c1.engine_model("e160"))   # from Engine
print(c1.info("Corolla", 1997))  # from Car
```

**Multiple inheritance** — one class inheriting from two parents:

```python
class Calc1:
    def subtract(self, a, b):
        return a - b

class Calc2:
    def multiply(self, c, d):
        return c * d

class Combined(Calc1, Calc2):   # inherits both
    def add(self, c, d):
        return c + d

com = Combined()
print(com.add(10, 12))
print(com.multiply(10, 12))
print(com.subtract(10, 12))
```

#### `super()` — calling the parent class from a child

```python
class Car:
    def __init__(self, name, brand):
        self.name = name
        self.brand = brand

    def show(self):
        print(self.name, self.brand)

class ElectricCar(Car):
    def __init__(self, model, price, name, brand):
        super().__init__(name, brand)   # call Car's __init__
        self.model = model
        self.price = price

    def show_all(self):
        return self.model, self.price, self.name, self.brand

e1 = ElectricCar("Model S", 8000000, "Tesla", "Tesla Motors")
e1.show()
e1.show_all()
```

#### Encapsulation — private attributes with `__`
Adding `__` before an attribute name makes it private — not accessible from outside the class.

```python
class Car:
    def __init__(self, name, year):
        self.name = name
        self.__year = year    # private — double underscore

    def show(self):
        print(self.name, self.__year)  # accessible inside the class

c1 = Car("Toyota", 2020)
c1.show()   # works fine
```

```python
# Trying to access __year directly from outside raises AttributeError
# c1.__year  -> AttributeError: 'Car' object has no attribute '__year'
print("Direct access blocked — use a method instead")
```

The right way — expose private data through a getter method:

```python
class Player:
    def __init__(self, name, age):
        self.name = name
        self.__age = age      # private

    def get_age(self):
        return self.__age     # controlled access via method

p1 = Player("Virat", 35)
print(p1.get_age())
```

Without `__init__`, Python prints the object's memory address by default:

```python
class Car:
    def __init__(self, name, year):
        self.name = name
        self.__year = year

c1 = Car("Python", 1991)
print(c1)   # shows memory address like <__main__.Car object at 0x...>
```

#### Nested (inner) classes
A class defined inside another class.

```python
class Outer:
    def __init__(self):
        self.name = "Outer"

    class Inner:
        def __init__(self):
            self.name = "Inner"

        def display(self):
            print("Hello from inner class")

outer = Outer()
inner = outer.Inner()   # access Inner through the outer instance
inner.display()
```

Inner class with a reference back to Outer:

```python
class Outer:
    def __init__(self):
        self.name = "Outer"

    class Inner:
        def __init__(self, outer_ref):
            self.name = "Inner"
            self.outer = outer_ref   # store reference to the outer class

        def display(self):
            print("Hello from inner class")

outer = Outer()
inner = outer.Inner(Outer)
inner.display()
```

A more practical nested class example — `Car` containing its own `Engine` class:

```python
class Car:
    def __init__(self, name, year):
        self.name = name
        self.year = year

    def display(self):
        return self.name, self.year

    class Engine:
        def __init__(self, model, hp):
            self.model = model
            self.hp = hp

        def show(self):
            return self.model, self.hp

my_car = Car("Toyota", 2020)
engine = my_car.Engine("V6", 250)
print(my_car.display())
print(engine.show())
```
