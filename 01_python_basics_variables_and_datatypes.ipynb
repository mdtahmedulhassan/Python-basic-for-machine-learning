#### 3. Variables
A variable is just a name that points to a value in memory.
We use `=` to assign a value to a variable.

```python
a = 10
a  # shows the value of a
```

```python
a
```

Variables can be **reassigned** any time. Python doesn't lock the type.

```python
a = 5
a
```

#### 4. Variable naming rules
- Names **cannot start with a number** (`9a = 80` → error)
- Names **can start with letters or underscore**, and can contain numbers after that
- Names are case-sensitive (`a` and `A` are different variables)

Below are a few examples showing what works and what doesn't.

❌ This will throw an error because a variable name can't start with a digit:

```python
 9a = 80   # SyntaxError: invalid decimal literal
9a
```

✅ But a number can appear later in the name:

```python
a9 = 90
a9
```

✅ Names can start with an underscore:

```python
_a = "Shadab"
_a
```

✅ Underscore can also be used in the middle of a name:

```python
Sha_dab = "profession"
Sha_dab
```

#### 5. Multiple assignment
You can assign the same value to multiple variables in one line.

```python
var1 = var2 = 10
print(var1)
print(var2)
```

#### 6. Unpacking values
You can assign multiple values to multiple variables in one line.
But the number of variables and values **must match**, otherwise Python raises a `ValueError`.

❌ Wrong number of values → error:

```python
a, b, c = 11, 5   # ValueError: not enough values to unpack (expected 3, got 2)
```

✅ Correct unpacking:

```python
a, b, c = 11, 5, 14
print("value of a is", a)
print("value of b is", b)
print("value of c is", c)
```

```python
a + b + c  # 11 + 5 + 14 = 30
```

#### 7. The `del` keyword
`del` removes a variable completely. After deleting, the variable no longer exists.

```python
d = b + c
d
```

```python
del b
# Now "b" no longer exists, but "d" still holds its old value (19)
print(b)
print(d)
```

#### 9. Data types
Python automatically figures out the data type based on the value you give.
Common basic types: `int`, `float`, `str`, `bool`

```python
a = 10
print(type(a))   # int -> integer number
```

```python
c = 10.5
print(type(c))   # float -> decimal number
```

```python
d = "Cricket"
print(type(d))   # str -> text
```

```python
d = True
print(type(d))   # bool -> True / False
```

#### 10. String indexing and slicing
Strings work like a sequence of characters, so we can access them by position (index).
- Index starts from **0**
- Negative index counts from the end (`-1` is the last character)

```python
d = "Sangakkara"
"p" in d  # checking if a character exists in the string
```

```python
d.find("g")   # returns the index of first occurrence of "m"
```

```python
d.index("g")  # returns the index of "T"
```

```python
d[-1]   # last character
```

```python
d[0]    # first character
```

##### Slicing syntax: `string[start : stop : step]`
- If start/stop is left empty, it means "from the beginning" / "till the end"
- step controls the direction and jump size

```python
d[::]      # full string (copy)
```

```python
d[::-1]    # reversed string
```

```python
d[:4]      # first 4 characters
```

```python
d[::2]     # every 2nd character
```

```python
d[-1:-7:-2]  # start from last, go backwards, take every 2nd char
```

##### 11. `range()`
`range(start, stop)` generates a sequence of numbers from start to stop-1.
It doesn't create a list immediately — it's a memory-efficient object.

```python
a = range(10, 100)
a
```

#### 12. Arithmetic operators
| Operator | Meaning |
|---|---|
| `+` | addition |
| `-` | subtraction |
| `**` | power (exponent) |
| `%` | modulus (remainder) |
| `//` | floor division (integer division) |

```python
a = 2 ** 4   # 2 to the power 4 = 16
a
```

```python
a % 2   # remainder when 16 is divided by 2
```

```python
15 % 2   # remainder of 15/2 -> 1
```

```python
15 // 2  # floor division -> 7 (drops the decimal part)
```

##### 13. `round()` function
Rounds a number to the given number of decimal places.

```python
print(round(3.1416, 2))  # rounds to 2 decimal places
```

##### 14. Augmented assignment operators
A shortcut for updating a variable based on its own value.
`a += 5` is the same as `a = a + 5`

```python
a = 10
a += 5  # same as a = a + 5
a
```
