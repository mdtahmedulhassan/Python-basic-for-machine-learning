### Python Functions

Functions let you reuse a block of code by giving it a name and calling it whenever needed.

Topics:
- defining and calling functions (`def`, `return`)
- parameters and arguments
- local vs global variables
- keyword arguments and default values
- `*args` and `**kwargs`
- calling one function from another
- practical examples: prime checker, even filter, calculator

#### 1. Built-in functions (quick reminder)
Python has many ready-made functions. We've been using these all along:

```python
# some common built-ins
print(10, 15, "Shadab")
print(len([1, 2, 3]))
print(sum([10, 15]))
print(type(3.14))
print(round(3.1416, 2))
```

#### 2. Defining your own function
Use `def` to create a function. Call it by writing its name followed by `()`.

```python
a = [1, 2, 5, 74, 8]
total = 0
for i in a:
    total = total + i
total  # doing this every time gets repetitive...
```

```python
# wrap it in a function so you can reuse it
def addit():
    a = [1, 2, 5, 74, 8]
    total = 0
    for i in a:
        total = total + i
    return total

addit()  # call it
```

#### 3. `return` vs `print` inside a function
`print` just displays something. `return` sends a value back to the caller so you can use it.

Anything after `return` is never reached — the function stops there.

```python
def greet():
    print("I love hilsa fish")  # this prints
    return "I love playing badminton"  # this is returned

greet()
```

```python
# once return is hit, everything after is ignored
def greet():
    return "I love playing badminton"
    print("this line never runs")

greet()
```

#### 4. Parametrised functions
Parameters are the variable names inside `def`. Arguments are the values you pass when calling.

```python
def name(a):  # "a" is the parameter
    return a

name("Sangakkara")  # "Sangakkara" is the argument
```

```python
def name(a, b):
    return a, b

name("Lara", "Shadab")
```

```python
def add(a, b):
    return a + b

add(10, 15)
```

Passing a list as an argument:

```python
def addit(a):
    total = 0
    for i in a:
        total = total + i
    return total

a = [1, 2, 5, 74, 8]
addit(a)
```

#### 5. Even number filter (function version)

```python
def even():
    a = [11, 2, 51, 74, 81]
    for i in a:
        if i % 2 == 0:
            print(i)

even()
```

#### 6. Prime number checker

```python
# check if a single number is prime
a = 11
for i in range(2, a):
    if a % i == 0:
        print("not prime")
        break
else:
    # the "else" on a for loop runs only if the loop wasn't broken
    print("prime")
```

```python
# wrap it in a function — input commented out, uses hardcoded value
def prime_check():
    a = 11  # try: a = int(input())
    for i in range(2, a):
        if a % i == 0:
            print("not prime")
            break
    else:
        print("prime")

prime_check()
```

```python
# find primes in a range
def primes_in_range():
    for i in range(11, 20):
        for j in range(2, i):
            if i % j == 0:
                print("not prime", i)
                break
        else:
            print("prime", i)

primes_in_range()
```

#### 7. Local vs Global variables
A variable defined inside a function is **local** — it only exists inside that function.
A variable defined outside is **global** — visible everywhere.

```python
def local_demo():
    x = 10  # local variable — only exists inside this function
    y = 15
    return x

local_demo()
```

```python
# x doesn't exist outside the function
# x   # this would give NameError: name 'x' is not defined
```

```python
x = 10
y = 15

def global_demo():
    return y + x  # can read global variables from inside a function

global_demo()
```

```python
y + x  # also works outside, of course
```

#### 8. f-string return values

```python
def info(name, age):
    return f"name is {name} and age is {age}"

info("Virat", 35)
```

Arguments are matched by **position** — order matters when not using keyword names:

```python
info(35, "Virat")  # age goes into "name", name into "age" — wrong order!
```

#### 9. Keyword arguments
Pass arguments by name — order no longer matters.

```python
def info(name, age):
    return f"name is {name} and age is {age}"

info(age=35, name="Virat")  # keyword args, order doesn't matter
```

#### 10. Default argument values
If an argument isn't passed, Python uses the default value.

```python
def info(name="Lara", age=40):
    return f"name is {name} and age is {age}"

info("Sangakkara")  # only name is passed, age uses default
```

#### 11. Calling functions from other functions

```python
def user1():
    a = 10  # simulating input
    b = 15
    return a + b

def user2():
    c = 3
    d = 5
    result = c * d
    u = user1()  # calling another function
    return result + u

user2()
```

---
### Summary
- `def` defines a function; `return` sends back the result.
- Parameters go in `def name(param1, param2)`. Arguments go in `name(val1, val2)`.
- Local variables only exist inside the function. Global variables are accessible everywhere.
- Keyword arguments let you pass values in any order.
- Default values make some arguments optional.
- Functions can call other functions.
