### Python Type Casting & Core Data Structures

In this notebook we'll cover:
- Type casting (converting between int, float, str)
- Taking input from the user
- Lists
- Tuples
- Sets
- Dictionaries

These are the most-used data structures in Python and you'll use them constantly in real projects.

#### 1. Checking and converting types
`type()` tells you what type a value is.
`int()`, `float()`, `str()` convert (cast) a value from one type to another.

```python
a = "10"
b = 10
c = True
type(b)
```

```python
a = 10
a + 2  # int + int = int
```

```python
b
```

#### Casting int to float

```python
a = 10
b = float(a)
print(type(a))   # original variable type stays int
print(b)          # but b is now a float
```

```python
float(5)  # converts int 5 to float 5.0
```

#### Casting float to int — decimal part is dropped (not rounded)

```python
v = int(1.9)
print(v)  # becomes 1, not 2
```

#### Casting number to string

```python
a = 10
str(a)
```

```python
print(type(str(a)))
```

```python
a = int(10.5)
a  # decimal part removed -> 10
```

```python
a = str(10.5)
a  # number becomes text "10.5"
```

#### Data Structures

Python has 4 main built-in collection types:
| Type | Ordered? | Changeable (mutable)? | Allows duplicates? | Syntax |
|---|---|---|---|---|
| List | Yes | Yes | Yes | `[ ]` |
| Tuple | Yes | No | Yes | `( )` |
| Set | No | Yes | No | `{ }` |
| Dictionary | Yes  | Yes | Keys: No | `{key: value}` |

## 4. List
A list is an ordered, changeable collection. It can hold mixed data types — numbers, strings, booleans, even other lists.

```python
a = [1, 2, "Ronaldo", 10.5, True]
print(a)
print(type(a))
```

```python
s = list([2, -2.14, "Messi", -99.999, False])  # creating a list using list()
s
```

Lists can contain other lists too (nested lists):

```python
s = list([1, 2, "Hello", ["biriani", "mutton", "illish"], 10.5, True])
s
```

```python
s[2]   # access element at index 2
```

```python
1 in s   # check if 1 exists in the list
```

```python
s[3][0]   # first item of the nested list at index 3
```

#### 5. Tuple
A tuple is just like a list, but **immutable** — once created, you cannot change its values.
Tuples use `()` instead of `[]`.

```python
a = (1, 2, 5.3, "virat", False, True)
a
```

```python
type(a)
```

Tuples can also contain nested tuples:

```python
a = (1, 2, 5.3, ("kohli", False), True)
a
```

An empty tuple:

```python
Shakib = ()
type(Shakib)
```

Creating a tuple from nested structures using `tuple()`:

```python
z = tuple((((15, 10), [10, 15]), 10))
z
```

Creating a tuple from a list:

```python
z = tuple([15, 10])
type(z)
```

```python
z
```

A **single-element tuple** needs a trailing comma — without it, Python won't treat it as a tuple.

```python
p = (9,)   # note the comma! without it, this would just be the number 9
p
```

```python
type(p)
```

#### 6. Set
A set is an **unordered** collection with **no duplicate values**.
Important: all items inside a set must be "hashable" — so lists (which are mutable) **cannot** go inside a set.

❌ This fails because a list is unhashable:

```python
a = {1, 2, 5.3, ("sajib", False), ["joshim", "jamboo"], True} # TypeError: unhashable type: \'list\'
```

✅ Tuples are fine inside a set (since tuples are immutable):

```python
a = {1, 2, 5, ("dildar", "bobita", "champa"), 15.5}
a
```

Creating a set from a list — notice duplicates get removed automatically:

```python
a = set([1, 2, 50])
a
```

#### 7. Dictionary
A dictionary stores data as **key-value pairs**. Keys must be unique, but values can be anything (including lists, tuples, etc).

```python
dic = {}
type(dic)   # empty dictionary
```

```python
dic = {"name": ("Rohan", "Joy", "Rakib"),
       "age": [20, 40]}
dic
```

Accessing a value by key, then indexing into it:

```python
dic["name"][1]   # get the "name" tuple, then its 2nd element
```

Getting all keys and all values:

```python
dic.keys()
```

```python
dic.values()
```

```python
type(dic)
```

---
### Summary
- Use `int()`, `float()`, `str()` to convert between types.
- `input()` always gives back a string — convert it if you need a number.
- **List** `[]` → ordered & changeable, allows duplicates.
- **Tuple** `()` → ordered & fixed (immutable), allows duplicates.
- **Set** `{}` → unordered, no duplicates, items must be hashable.
- **Dictionary** `{key: value}` → key-value pairs, great for labeled data.
