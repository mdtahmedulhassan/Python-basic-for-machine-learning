#### 3. `*args` — variable-length arguments
`*args` collects all positional arguments into a tuple. You can pass as many values as you like.

```python
def addi(*a):
    return sum(a)  # a is a tuple of all passed values

addi(10, 15, 20, 40)
```

⚠️ Passing a list directly to a `*args` function doesn't work the way you'd expect:

```python
a = [10, 15, 12, 17, 16, 18]
addi(a)  # TypeError: sum gets a tuple containing one list, not numbers
# Fix: pass the list normally (without *args), or unpack it with addi(*a)
```

```python
# correct way when you already have a list
def addi(a):
    return sum(a)

a = [10, 15, 12, 17, 16, 18]
addi(a)
```

#### 4. `**kwargs` — variable-length keyword arguments
`**kwargs` collects keyword arguments into a **dictionary**.

```python
def player_info(**info):
    return info

player_info(name="Sangakkara", country="Sri Lanka", runs=14000)
```

#### 5. Iterating a dictionary with `.items()`

```python
dic = {
    "name": ["Sangakkara", "Lara"],
    "age": [46, 55],
    "profession": ["cricketer", "cricketer"]
}
```

```python
dic
```

```python
# access values by key
dic["name"]
```

```python
dic["name"][0]
```

```python
# loop over all key-value pairs
def show_info(info):
    for key, value in info.items():
        print(key, "->", value)

show_info(dic)
```

#### 6. Simple calculator using functions

```python
def cal():
    print("welcome to my calculator")
    print("1. Addition")
    print("2. Subtraction")
    print("3. Multiplication")
    print("4. Division")

    choice = "1"  # simulating input — try: choice = input("enter 1/2/3/4: ")
    num1, num2 = 10.0, 5.0  # try with actual input()

    if choice == "1":
        return num1 + num2
    elif choice == "2":
        return num1 - num2
    elif choice == "3":
        return num1 * num2
    elif choice == "4":
        if num2 != 0:
            return num1 / num2
        else:
            print("cannot divide by zero")
    else:
        print("invalid choice — pick 1, 2, 3, or 4")

cal()
```

## Exception Handling — `try / except / finally`

Errors (exceptions) happen. Instead of letting your program crash, you can **catch** the error and handle it nicely.

#### 7. ZeroDivisionError — without try/except

```python
a = 10
res = a / 0
print("hello")
# ZeroDivisionError: division by zero — program crashes here, "hello" never prints
```

#### 8. Using `try / except`
Code inside `try` runs normally. If an error happens, Python jumps to `except`.

```python
try:
    a = 10
    res = a / 0
except ZeroDivisionError:
    print("cannot divide a number by 0")
```

#### 9. Catching multiple exceptions

```python
try:
    a = 10
    b = 0  # try: b = int(input())
    res = a / b
except TypeError:
    print("cannot divide a number by a string")
except ZeroDivisionError:
    print("do not use 0 as the divisor")
```

#### 11. `finally` block
`finally` always runs, whether an error happened or not — good for cleanup steps.

```python
try:
    dic = {"name": "Shadab", "age": 25}
    dic["nonexistent_key"]  # KeyError
except (NameError, KeyError):
    print("key not found in dictionary")
finally:
    print(dic["age"])  # this always runs
```

#### 12. Generic `Exception` — catch any error
`Exception as error` gives you the actual error message.

```python
try:
    a = 10
    b = 0  # try: b = int(input())
    c = a / b
    print(c)
except Exception as error:
    print(f"error happened: {error}")
```
