### Assignment 3 — Calculator with Exception Handling + Star Patterns

This notebook has two parts:
1. A fully working calculator using `while True` loop + `try/except` for error handling
2. Star pattern problems using nested loops

#### Part 1 — Calculator with Exception Handling

The calculator:
- Shows a menu (add, subtract, multiply, divide, exit)
- Keeps running in a loop until the user presses 5 to exit
- Handles invalid input (letters instead of numbers) with `try/except`
- Handles division by zero separately

```python
def calculator():
    print("Press 1 to perform addition")
    print("Press 2 to perform subtraction")
    print("Press 3 to perform multiplication")
    print("Press 4 to perform division")
    print("Press 5 to exit")

    while True:
        try:
            choice = int(input("Enter your choice: "))

            if choice == 5:
                print("Calculator turned off")
                break  # exits the while loop

            val1 = float(input("Enter first value: "))
            val2 = float(input("Enter second value: "))

            if choice == 1:
                res = val1 + val2
                print(f"Addition of {val1} and {val2} is {res}")
            elif choice == 2:
                res = val1 - val2
                print(f"Subtraction of {val1} and {val2} is {res}")
            elif choice == 3:
                res = val1 * val2
                print(f"Multiplication of {val1} and {val2} is {res}")
            elif choice == 4:
                try:
                    res = val1 / val2
                    print(f"Division of {val1} and {val2} is {res}")
                except ZeroDivisionError as er:
                    print(f"Division can't be performed: {er}")
            else:
                print("Invalid choice. Please enter [1/2/3/4/5]")

        except ValueError:
            # catches when someone types a letter instead of a number
            print("Please enter a valid number [1/2/3/4/5]")
```

```python
calculator()
```

#### Part 2 — Star Pattern Problems

These are classic nested loop problems. A good way to practice `for` loops and `print()` tricks like `end=`.

**Pattern 1** — Right-aligned triangle (grows right from the bottom)

```python
n = int(input("Enter n: "))
for i in range(n):
    for j in range(n - i - 1):
        print(" ", end="")   # leading spaces
    for j in range(i + 1):
        print("*", end=" ")  # stars
    print()
```

**Pattern 2** — Diamond (upside-down triangle + right-aligned triangle)

```python
n = int(input("Enter n: "))
# top half — shrinking
for i in range(n):
    for j in range(i):
        print(" ", end="")
    for j in range(n - i):
        print("*", end=" ")
    print()
# bottom half — growing
for i in range(n):
    for j in range(n - i - 1):
        print(" ", end="")
    for j in range(i + 1):
        print("*", end=" ")
    print()
```

**Pattern 3** — Right-aligned triangle with wider spacing

```python
n = int(input("Enter n: "))
for i in range(n):
    for j in range(n - i - 1):
        print(" ", end=" ")   # double space for wider alignment
    for j in range(i + 1):
        print("*", end=" ")
    print()
```

**Pattern 4** — Triangle + inverted triangle (diamond shape)

```python
n = int(input("Enter n: "))
# growing triangle
for i in range(n):
    for j in range(n - i - 1):
        print(" ", end="")
    for j in range(i + 1):
        print("*", end=" ")
    print()
# shrinking triangle
for i in range(n):
    for j in range(i + 1):
        print(" ", end="")
    for j in range(n - i - 1):
        print("*", end=" ")
    print()
```
