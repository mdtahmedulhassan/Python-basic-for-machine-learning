### Python Lists — Methods Deep Dive (+ a bit of Tuples)

This notebook focuses on the **methods** available on lists:
- adding items: `append()`, `extend()`, `insert()`
- removing items: `pop()`, `remove()`, `clear()`, `del`
- sorting: `sort()`, `sorted()`, `reverse()`
- copying lists correctly (`copy()` vs reference assignment)
- other useful methods: `count()`, `index()`, list slicing assignment
- a quick look at tuples: unpacking, repetition, and immutability

#### 1. Creating empty lists
There are two ways to create an empty list:

```python
li = []
li
```

```python
a = list()
a
```

##### 2. `append()` vs `extend()` vs `insert()`
- `append(x)` → adds `x` as a **single item** at the end (even if `x` is a list)
- `extend(iterable)` → adds each item from the iterable **individually**
- `insert(index, x)` → inserts `x` at a specific position

```python
li.append(1)
li.append(1)
li.append(1)
print(li)
```

Appending a list adds it as ONE nested item:

```python
li.append([10, 15])
li
```

Extending with a string adds each **character** as a separate item:

```python
li.extend("sji")
li
```

Extending with a list adds each element individually:

```python
li.extend([10.0])
li
```

```python
li.extend([10.0, "Sangakkara"])
li
```

##### 3. `insert(index, value)`
Inserts a value at the given index, shifting later elements to the right.

```python
a = [1, 5, 8, 9]
a
```

```python
a.insert(2, ["Lara", 10])
a
```

---
##### 4. Removing items: `del`, `pop()`, `remove()`, `clear()`

| Method | What it does |
|---|---|
| `del list[i]` | deletes item at index `i` (or a slice) |
| `list.pop()` | removes & returns the **last** item (or item at given index) |
| `list.remove(value)` | removes the **first** matching value |
| `list.clear()` | removes everything, leaves an empty list |
| `del list` | deletes the variable itself |

`del` can remove the whole variable — after this, the name no longer exists:

```python
# del(li)
# li   # NameError: name 'li' is not defined, because we deleted it above
```

```python
a = [1, 5, ["Lara", 10], "Lara", 8, 9]
a.pop()   # removes & returns the LAST item
```

```python
a
```

`remove(value)` deletes the first item that matches the given value:

```python
a.remove(5)
a
```

```python
a.remove(["Lara", 10])
a
```

```python
a.extend([10, 15, "Sangakkara"])
a
```

`remove()` only removes the **first** matching occurrence:

```python
a.remove("Lara")
a
```

```python
a.clear()
a   # now completely empty
```

##### 5. `pop(index)` — remove item at a specific position

```python
a = [1, 5, 8, 9]
a.pop(2)   # removes & returns item at index 2 (value 8)
```

```python
a
```

`del` also works with slices:

```python
a = [1, 5, 8, 9, 15, 12, 17, 18]
del a[2]
a
```

```python
del a[2:4]
a
```

##### 6. Sorting: `sort()` vs `sorted()`
- `list.sort()` sorts the list **in place** (changes the original, returns `None`)
- `sorted(list)` returns a **new sorted list**, leaving the original unchanged

```python
a = [1, 15, 9, 12, 8]
a.sort()
a
```

```python
a = [1, 15, 9, 12, 8]
a.sort(reverse=True)
a
```

⚠️ Sorting a list with **mixed types** (numbers and strings together) raises an error, because Python can't compare an `int` with a `str`:

```python
# text = ["Sangakkara", 9, "anaconda", 15, "wednesday", 7, "breaking", 5]
# text.sort()
# TypeError: '<' not supported between instances of 'int' and 'str'
```

Sorting works fine when all items are the same type:

```python
text = ["Sangakkara", "anaconda", "wednesday", "breaking"]
text.sort()
text
```

`sorted()` doesn't change the original list — it gives you a sorted copy:

```python
a = [1, 15, 9, 12, 8]
b = sorted(a, reverse=True)
b
```

Sorting strings is case-sensitive — uppercase letters come before lowercase ones in ASCII order:

```python
thislist = ["banana", "Orange", "Kiwi", "cherry"]
thislist.sort()
print(thislist)
```

##### 7. `reverse()` — reverses the list in place

```python
a = [1, 15, 9, 12, 8]
a.reverse()
a
```

#### 8. Copying a list properly
⚠️ Important concept: if you do `x = a`, `x` is just **another name for the same list** —
changing `x` also changes `a`! Use `.copy()` to make an independent copy.

```python
a
```

`.copy()` creates a real, independent copy:

```python
d = a.copy()
d
```

```python
d.append(10)
d   # only d changes
```

```python
a   # a stays the same, proving d is a separate list
```

But plain assignment (`x = a`) does NOT copy — both names point to the same list:

```python
x = a
x
```

```python
x.pop()   # modifying x...
```

```python
x   # x changed
```

```python
a   # ...and "a" changed too! Because x and a are the SAME list in memory
```

### A quick look at Tuples

Tuples are like lists, but **immutable** (can't be changed after creation).

```python
a = (1, 2, 5, 8, 9)
a
```

```python
print(type(a))
```

```python
b = tuple()
type(b)
```

Tuples don't support methods that modify them, like `insert()`:

```python
# b.insert(10)
# AttributeError: 'tuple' object has no attribute 'insert'
# Tuples are immutable, so there's no insert/append/remove for them.
```

#### `+=` style trick for tuples
You can't directly modify a tuple, but you CAN convert it to a list, change it, and convert it back.

```python
b = ("Lara", "Lara")
b = list(b)
b.append("Virat")
b = tuple(b)
b
```

```python
"Virat" in b
```

Indexing and slicing work the same way as lists:

```python
a = (10, 15, 14, 19, 17, 8)
a[2]
```

```python
a[-2]
```

```python
a[-2:-6:-1]
```

```python
a   # tuple stays unchanged, slicing just returns a new tuple
```

#### Tuple unpacking
You can assign all elements of a tuple to variables in one line.

```python
x, y, z, t, w, v = a
x
```

#### Unpacking with `*` (star expression)
`*z` collects all the "remaining" values into a list.

```python
x, y, *z = a
x
```

```python
y
```

```python
z   # everything else, collected into a list
```

#### Repeating a tuple with `*`

```python
fruits = ("apple", "banana", "cherry")
mytuple = fruits * 2
mytuple
```

```python
mytuple = fruits * 4
mytuple
```

---
### Summary
- `append()` adds ONE item; `extend()` adds items individually; `insert()` places at a specific index.
- `pop()`, `remove()`, `clear()`, and `del` all remove items but behave differently.
- `sort()` changes the list in place; `sorted()` returns a new sorted list.
- `x = a` does **not** copy a list — both names point to the same object. Use `.copy()` for a real copy.
- Tuples are immutable — no `append`/`insert`/`remove`, but you can unpack, slice, and repeat them with `*`.
