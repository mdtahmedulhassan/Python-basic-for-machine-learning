# Python Lists — Indexing, Slicing & Common Methods

Lists are one of the most used data structures in Python. In this notebook we'll go deeper into:
- indexing and slicing (including nested lists)
- common list methods: `append()`, `extend()`, `insert()`

This builds on the basic list intro from the previous notebook.

## 1. Creating a list with mixed data types

```python
a = [1, 2, 58, 7, 96, 10, "sajib"]
a
```

## 2. Indexing
Remember: indexing starts from 0.

```python
a[0]   # first element
```

## 3. Slicing
`list[start:stop]` gives elements from `start` up to (but not including) `stop`.

```python
a[1:3]   # elements at index 1 and 2
```

```python
len(a)   # number of elements in the list
```

```python
type(a)
```

`a[::]` with nothing on either side just returns a full copy of the list.

```python
a[::]
```

Using a step of `-1` reverses the list.

```python
a[::-1]   # reversed list
```

```python
a[:5]   # first 5 elements
```

```python
a[5:]   # from index 5 to the end
```

Negative slicing — start from the end and move backwards:

```python
a[-1:-6:-1]   # last element going backwards, 5 elements
```

```python
a[5]   # element at index 5
```

## 4. Nested lists
A list can contain other lists. To access an item inside a nested list, use multiple indexes — one for each level.

```python
a = [1, 2, [58, 7], [96, 10, "sajib"]]
a
```

```python
a[2][1]   # 2nd item -> the list [58, 7] -> its index 1 -> 7
```

Deeper nesting works the same way — just keep chaining indexes:

```python
a = [12, 25, [14, 5, 48, [1, 5, 8, [1, 4, 7, ["sajib", "45"]]]]]
a
```

```python
a[2][2]   # index into the nested list step by step
```

```python
a[2][3][3][1]   # going 4 levels deep!
```

## 5. Common list methods

### `append()` — add one item to the end of the list

```python
a = []
a.append(10)
a
```

```python
a.append("rahul")
a
```

You can even append a list or tuple as a single item:

```python
a.append(["sakib", "sadia"])
a
```

```python
a.append(("krish", "priya"))
a
```

### `extend()` — add multiple items from another list (unpacks them one by one)
Note the difference from `append()`: `append` adds the whole list as ONE item,
but `extend` adds each item separately.

```python
a.extend([10])
a
```

```python
a.extend([15, 20])
a
```

### `insert(index, value)` — insert a value at a specific position
Everything after that position shifts to the right.

```python
a.insert(1, 100)   # insert 100 at index 1
a
```

```python
a.insert(1, [10, 15])   # you can insert a list too, as a single item
a
```

---
### Summary
- `list[start:stop:step]` is the general slicing pattern — works on lists and strings the same way.
- Negative indexes/steps let you read a list backwards.
- For nested lists, chain indexes (`a[2][3][1]`) to dig deeper.
- `append()` adds ONE item (even if that item is a list/tuple).
- `extend()` adds items from another iterable individually.
- `insert(index, value)` places a value at a specific position, shifting others right.
