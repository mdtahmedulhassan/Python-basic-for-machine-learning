### Assignment — Conditional Statements (if / elif / else)

This notebook is a set of small practice problems on Python's `if`, `elif`, and `else`
statements, plus comparison and logical operators (`and`, `or`).

For each question, the code takes input from the user, so when you run a cell
you'll be asked to type a value in the box that appears.

#### Question 1
Take a number from the user and check:
- if it's an integer, check whether it's **even** or **odd**
- if it's not a whole number (has decimals), print that it's neither odd nor even

```python
x = float(input("Enter a number: "))

if x.is_integer():
    x = int(x)
    if x % 2 == 0:
        print('The number is even.')
    else:
        print('The number is odd.')
else:
    print('It is neither odd nor even')
```

#### Question 2
Take a number from the user and check whether it is **positive**, **negative**, or **zero**.

```python
x = float(input("Enter a number: "))

if x == 0:
    print('The number is zero')
elif x < 0:
    print('The number is negative')
else:
    print('The number is positive')
```

#### Question 3
Take a single character from the user and check whether it is a **vowel**, a **consonant**,
or not a letter at all.

```python
x = input("Enter a character: ")
vowels = ['a', 'e', 'i', 'o', 'u', 'A', 'E', 'I', 'O', 'U']

if x in vowels:
    print('Vowel')
elif x.isalpha():
    print('Consonant')
else:
    print('Not a letter')
```

#### Question 4
Take a year from the user and check whether it is a **leap year**.

Leap year rule: a year is a leap year if it's divisible by 4,
**except** centuries (divisible by 100) — unless they're also divisible by 400.

```python
x = int(input("Enter a year: "))

if (x % 4 == 0 and x % 100 != 0) or (x % 400 == 0):
    print(f"The year is {x} and it's a leap year")
else:
    print(f"The year is {x} and it's not a leap year")
```

#### Question 5
Take a score (0-100) and print the corresponding grade based on this scale:

| Score range | Grade |
|---|---|
| 0 - 39 | F |
| 40 - 59 | D |
| 60 - 69 | C |
| 70 - 79 | B |
| 80 - 100 | A |
| anything else | Not in grading system |

```python
x = float(input("Enter your score: "))
print(f"The score is {x}")

if 0 <= x < 40:
    print('F')
elif 40 <= x < 60:
    print('D')
elif 60 <= x < 70:
    print('C')
elif 70 <= x < 80:
    print('B')
elif 80 <= x <= 100:
    print('A')
else:
    print('Not in grading system')
```

#### Question 6
Take a number and check whether it is divisible by **both 5 and 11**.

```python
x = float(input("Enter a number: "))

if x % 11 == 0 and x % 5 == 0:
    print(x, 'is divisible by both 5 and 11')
else:
    print(x, 'is not divisible by both 5 and 11')
```

#### Question 7
Take **three numbers** from the user and find the largest one.

```python
a = float(input("Enter first number: "))
b = float(input("Enter second number: "))
c = float(input("Enter third number: "))

print('The numbers are', a, b, c)

if a >= b and a >= c:
    print(a, 'is the largest among the three numbers')
elif b >= a and b >= c:
    print(b, 'is the largest among the three numbers')
else:
    print(c, 'is the largest among the three numbers')
```

#### Question 8
A simple traffic-light simulator. Take a color name from the user
(red/yellow/green — case-insensitive) and print what it means.

```python
color = input("Enter traffic light color: ").lower()

if color == 'red':
    print('Stop')
elif color == 'yellow':
    print('Ready')
elif color == 'green':
    print('Go')
else:
    print('Not a valid traffic light color')
```

### Summary
- `if / elif / else` lets your program choose between different actions based on conditions.
- Comparison operators: `==`, `!=`, `<`, `>`, `<=`, `>=`
- Logical operators `and` / `or` let you combine multiple conditions.
- `.is_integer()` checks if a float has no decimal part.
- `.isalpha()` checks if a string contains only letters.
- Always remember to convert `input()` (which is always a string) using `int()` or `float()` when you need a number.
