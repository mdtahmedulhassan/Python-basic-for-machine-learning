### Python Strings — Formatting & Built-in Methods

Strings are everywhere in Python. In this notebook we'll cover:
- different ways to print/format strings (f-strings, `.format()`, multi-line strings)
- common string methods (case conversion, search, replace, split/join, strip, etc.)
- checking string properties (`isalpha`, `isdigit`, `isalnum`, etc.)

##### 1. Printing strings together with `print()`
You can pass multiple values to `print()` separated by commas — Python adds a space between them automatically.

```python
a = "i love ml"
b = "i love dl"

print("due to the current demand", a)
print("due to the advancement of research", b)
```

```python
print(a, "due to the current demand")
print(b, "due to the advancement of research")
```

```python
print("due to the", a, "current demand")
```

#### 2. f-strings (formatted string literals)
The cleanest way to insert variable values inside a string. Just put `f` before the quotes
and wrap variable names in `{}`.

```python
a = 10
b = 20
c = a + b
print(f'value of a is {a} and value of b is {b}, their sum is {c}')
```

#### 3. `.format()` method
An older alternative to f-strings — placeholders `{}` get filled in order with `.format()` arguments.

```python
print("value of a is {} and value of b is {}, their sum is {}".format(b, a, c))
```

f-strings also support special characters like `\\n` (newline) inside them.

```python
print(f'value of a is {a} \n and value of b is {b} \n their sum is {c}')
```

#### 4. Multi-line strings
Triple quotes (`"""` or `'''`) let a string span multiple lines.

```python
a = """i love ml.
   i love dl"""
a
```

#### 5. Case conversion methods
- `.capitalize()` → makes only the first letter uppercase
- `.upper()` → all uppercase
- `.lower()` → all lowercase

Note: these methods return a **new** string — they don't change the original string.

```python
b= "Who's gonna win?"
```

```python
b.capitalize()
```

```python
c = b.upper()
c
```

Original `b` stays unchanged — strings are immutable:

```python
b
```

```python
c
```

```python
c.lower()
```

#### 6. Length and counting

```python
len(c)
```

```python
a.count("l")   # how many times "l" appears
```

```python
a.find("l")   # index of first occurrence of "l"
```

##### 7. `split()` and `join()`
- `.split()` breaks a string into a **list of words** (splits on spaces by default).
- `" ".join(list)` does the opposite — joins a list of words back into a string.

```python
a = "i love watching interstellar movie"
a
```

```python
b = a.split()
b
```

```python
b = " ".join(b)
b
```

##### 8. `.strip()`
Removes extra whitespace from the beginning and end of a string.

```python
a = "  love watching     anaconda movie         "
a.strip()
```

#### 9. String concatenation with `+`
Joins two strings together directly.

```python
a = "Virat "
b = "Kohli"
a + b
# a + " " + b  -> adds a space between them
```

#### 10. Slicing strings
Same slicing rules as lists: `string[start:stop:step]`

```python
a = "i love watching inception movie"
a[::]   # full copy
```

```python
a[:5]
```

```python
a[2:5]
```

```python
a[::3]
```

```python
a[::-1]   # reversed string
```

```python
a[-5:-12:-2]
```

```python
a[-12:-5]
```

##### 11. `.replace()`
Returns a new string with all occurrences of a substring replaced. The original string is unchanged.

```python
a
```

```python
v = a.replace("anaconda", "tiger")
v
```

```python
a   # original string is still the same
```

```python
v   # new string with the replacement
```

##### 12. `.center()`
Centers the string within a given width, padding with a chosen character.

```python
a = "Shadab"
a.center(20, "0")
```

```python
a = "Shadab"
print(a)
```

##### 13. `.encode()`
Converts a string into bytes (binary format) — useful when working with files, networks, etc.

```python
c = a.encode()  # converts the string into binary (bytes) format
c
```

##### 14. `.endswith()`
Checks whether a string ends with a given substring (returns `True`/`False`).

```python
a.endswith("b")
```

```python
a.endswith("c")
```

##### 15. Checking string content — `is...()` methods
| Method | Checks if... |
|---|---|
| `isalnum()` | all characters are letters or digits |
| `isdigit()` | all characters are digits |
| `isalpha()` | all characters are letters |
| `islower()` | all characters are lowercase |

```python
a = "10"
a.isalnum()   # digits count as alphanumeric -> True
```

```python
a = "Shadab"
"i" in a   # check if a character exists in the string
```

```python
a.isdigit()   # "Shadab" is not all digits -> False
```

```python
a.isalpha()   # "Shadab" is all letters -> True
```

```python
a = "@"
a.isalnum()   # @ is a special character, not alphanumeric -> False
```

```python
a.islower()   # "@" has no case, so islower() is False
```

---
### Summary
- f-strings (`f'{var}'`) are the easiest way to insert variables into strings.
- `.split()` / `.join()` convert between strings and lists of words.
- `.strip()` removes leading/trailing whitespace.
- `.replace()` returns a new string — original is unchanged (strings are immutable).
- `is...()` methods are handy for validating user input.
