### Python Conditional Statements — if / elif / else

Making decisions in code. Python uses `if`, `elif`, and `else` to run different blocks based on conditions.

This notebook also has a quick intro to `for` and `while` loops at the end.

#### 1. Simple if / else
`if` checks a condition. If it's `True`, it runs the first block. Otherwise it runs `else`.

````python
a = 20

if a == 20:
    print("I can buy some mango")
else:
    print("I can buy some pineapples")
````

````python
a = 2

if a == 20:
    print("I can buy some mango")
else:
    print("I can buy some pineapples")
````

#### 2. if / elif / else
`elif` lets you check more conditions between `if` and `else`. Python checks them in order and stops at the first `True`.

````python
v = 18

if v == 20:
    print("buy mango")
elif v > 20:
    print("buy pineapple")
else:
    print("buy some banana")
````

#### 3. Using `and` / `or` in conditions
- `and` → both sides must be True
- `or` → at least one side must be True

````python
taka = 17

if taka > 15 or taka == 20:
    print("you can buy a book")
elif taka > 20 and taka <= 25:
    print("you can buy a pen")
elif taka > 25 and taka <= 30:
    print("buy a shirt")
elif taka > 30 and taka <= 40:
    print("buy some clothes")
else:
    print("buy fiction books")
````

#### 5. Grade checker example
Practical use of `elif` — checking ranges for grades.

```
< 40   → Fail
40–50  → B+
51–59  → A-
60–69  → A
70–79  → A+
80+    → Outstanding
```

````python
# grade checker
score = 75  # change this to test different scores

if score < 40:
    print("Fail")
elif 40 <= score <= 50:
    print("B+")
elif 51 <= score <= 59:
    print("A-")
elif 60 <= score <= 69:
    print("A")
elif 70 <= score <= 79:
    print("A+")
else:
    print("Outstanding")
````

#### 6. Quick intro to loops (covered more in the next notebook)

**`for` loop** — iterates over a sequence (list, string, range, etc.)

````python
a = "i love"
len(a)  # 6 characters
````

````python
for index in a:
    print(index)  # prints each character one by one
````

````python
for number in range(5, 10):
    print(number)  # 5 to 9
````

````python
for number in range(10, 5, -2):
    print(number)  # counts down: 10, 8, 6
````

Iterating a reversed string:

````python
a = "i love"
for value in a[::-1]:
    print(value)
````

#### 7. `continue` and `break`
- `continue` → skip this iteration, move to next
- `break` → stop the loop entirely

````python
# skip 20 and 25, print everything else 1-49
for i in range(1, 50):
    if i == 20 or i == 25:
        continue
    print(i)
````

````python
# stop as soon as we hit 10
for i in range(1, 50):
    if i == 10:
        break
    print(i)
````

**`while` loop** — keeps running as long as the condition is `True`.

````python
i = 1
while i <= 6:
    print(i)
    i = i + 1  # always update the counter or you get an infinite loop!
````

### Summary
- `if / elif / else` lets your program make decisions.
- Use `and`/`or` to combine multiple conditions.
- `continue` skips an iteration; `break` exits the loop completely.
- `while` loops run until the condition becomes `False`.
