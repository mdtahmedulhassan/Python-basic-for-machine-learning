### Python List Comprehension, `map()`, `filter()` & `lambda`

Three powerful ways to transform lists in Python — all doing the same job,
just with different styles. Also covers flattening nested lists.

#### 1. The long way — loop + `append()`

```python
a = [1, 2, 5, 8]
for i in a:
    print(i * i)  # print squares
```

#### 2. List comprehension — shorter, more Pythonic
Syntax: `[expression for item in iterable]`

```python
a = [1, 2, 5, 8]
res = [i * i for i in a]
res
```

```python
# squares of 0-9
res = [i * i for i in range(10)]
res
```

#### List comprehension with `if` condition

```python
a = [0, 1, 4, 9, 16, 25, 36, 49, 64, 81]

# loop version
for i in a:
    if i % 2 == 0:
        print("even")
    else:
        print("odd")
```

```python
# one-liner version
res = ["odd" if i % 2 != 0 else "even" for i in a]
res
```

Nested ternary — find largest among three values:

```python
a, b, c = 10, 12, 11

res = ["a is big" if a > b and a > c else ("b is big" if b > c and b > a else "c is big")]
res
```

Sum using walrus operator `:=` (Python 3.8+):

```python
a = [0, 1, 4, 9, 16, 25, 36, 49, 64, 81]

tot = 0
[tot := tot + i for i in a]
tot
```

#### 3. `filter()` — keep only items that pass a condition
Syntax: `filter(function, iterable)`

```python
a = [0, 1, 4, 9, 16, 25, 36, 49, 64, 81]

def is_even(x):
    return x % 2 == 0

evens = list(filter(is_even, a))
evens
```

Same with a `lambda`:

```python
evens = list(filter(lambda x: x % 2 == 0, range(10)))
evens
```

---
#### 4. `map()` — apply a function to every item
Syntax: `map(function, iterable)`

```python
a = [0, 1, 4, 9, 16, 25, 36, 49, 64, 81]

def square(x):
    return x ** 2

result = list(map(square, a))
result
```

`map()` returns True/False when the function is a condition:

```python
def is_even(x):
    return x % 2 == 0

result = list(map(is_even, a))
result
```

Same with `lambda`:

```python
sq = list(map(lambda y: y ** 2, range(10)))
sq
```

#### Comparing loop vs list comprehension vs `map()`
All three produce the same output — pick the one that reads clearest to you.

```python
a = ["a", "b", "c"]

# loop
b = []
for i in a:
    b.append(i.upper())
print(b)
```

```python
# list comprehension
upp = [i.upper() for i in a]
print(upp)
```

```python
# map + lambda
up1 = list(map(lambda i: i.upper(), a))
print(up1)
```

Extract first character of each word:

```python
a = ["Apple", "Ball", "Cat"]

# loop
n = []
for i in a:
    n.append(i[0])
print(n)
```

```python
# list comprehension
li = [i[0] for i in a]
li
```

```python
# map
lm = list(map(lambda i: i[0], a))
lm
```

#### Nested loops in list comprehension
Useful for working with nested lists (2D or 3D structures).

```python
# triple-nested loop — manual
a = [[[1, 2, 5, 8, 74, 6, 8]]]
for i in a:
    for j in i:
        for k in j:
            print(k)
```

```python
# same as list comprehension
res = [k for i in a for j in i for k in j]
res
```

#### Flattening a deeply nested list — recursive function
When nesting is unpredictable, a recursive approach handles any depth.

```python
def flatten(li):
    result = []
    for item in li:
        if isinstance(item, list):
            result.extend(flatten(item))  # go deeper
        else:
            result.append(item)
    return result

a = [[[1, 2, 5], [8, 74, 6, 8]], [[11, 12, 15], [81, 174, 16, 18]]]
flatten(a)
```

`map()` with `flatten` — flattens each sub-list:

```python
lam2 = list(map(flatten, a))
lam2
```
